```python
import requests
import time
import matplotlib.pyplot as plt

latitudes = []
longitudes = []
timestamps = []

print("Starting ISS tracker... This will take ~17 minutes.\n")

for i in range(100):
    try:
        response = requests.get("http://api.open-notify.org/iss-now.json")
        data = response.json()

        lat = float(data["iss_position"]["latitude"])
        lon = float(data["iss_position"]["longitude"])
        ts = data["timestamp"]

        latitudes.append(lat)
        longitudes.append(lon)
        timestamps.append(ts)

        print(f"[{i+1}/100] Timestamp: {ts} | Latitude: {lat:.2f}, Longitude: {lon:.2f}")
        time.sleep(10)
    except Exception as e:
        print("Error:", e)
        time.sleep(10)
        
plt.figure(figsize=(12, 6))
plt.plot(longitudes, latitudes, marker='o', markersize=3, linestyle='-', linewidth=1)
plt.title("ISS Path over Time")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
